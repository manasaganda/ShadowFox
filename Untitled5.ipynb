```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
```

```python
# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Example input text
input_text = "The future of artificial intelligence is"

# Tokenize the input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")
```

```python
input_text = "In a distant future, humans have colonized Mars. The challenges they face include"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

output_ids = model.generate(input_ids, max_length=60, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print("Generated Text:", output_text)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Example text generated
generated_texts = [tokenizer.decode(model.generate(tokenizer.encode("AI is", return_tensors="pt"), max_length=50)[0], skip_special_tokens=True) for _ in range(10)]

# Tokenize and count word frequencies
word_counts = Counter(" ".join(generated_texts).split())

# Plot the word frequencies
plt.figure(figsize=(10, 5))
plt.bar(*zip(*word_counts.most_common(10)))
plt.title("Top 10 Word Frequencies in Generated Text")
plt.show()
```
